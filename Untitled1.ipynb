```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

df = pd.read_csv('_cancer_dataset_uae.csv')
clean = df[['Age', 'Weight']].dropna()

# Pearson
r_p, p_p = pearsonr(clean['Weight'], clean['Age'])

# Spearman
r_s, p_s = spearmanr(clean['Weight'], clean['Age'])

print("Hypothesis Test Results: 'Age' (Dependent) vs 'Weight' (Independent)\n")
print("Pearson Test:")
print(f"  correlation: {r_p:.4f}")
print(f"  p_value: {p_p:.3f}")
print("  interpretation: No significant linear correlation (p  0.05)\n")

print("Spearman Test:")
print(f"  correlation: {r_s:.4f}")
print(f"  p_value: {p_s:.4f}")
print("  interpretation: No significant monotonic correlation (p  0.05)")
```

```python
import pandas as pd
from scipy.stats import ttest_1samp

df = pd.read_csv('_cancer_dataset_uae.csv')
sample = df['Age'].dropna()          # treat Age as the value variable
pop_mean = 50                        # hypothesised population mean (change if needed)

t_stat, p_val = ttest_1samp(sample, pop_mean)

print("One-Sample T-Test Results for 'value':")
print(f"Population Mean: {pop_mean}")
print(f"Sample Mean: {sample.mean():.3f}\n")

print("Test Metrics:")
print(f"  t_statistic: {t_stat:.3f}")
print(f"  p_value: {p_val:.3f}\n")

print("Interpretation:", end=" ")
if p_val < 0.05:
    print("Reject H  the sample mean differs significantly from the population mean.")
else:
    print("Fail to reject H  no significant difference from the population mean.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats # For Q-Q Plot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # Using Linear Regression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# --- 1. Load Data ---
FILE_PATH = '_cancer_dataset_uae.csv'
df = pd.read_csv(FILE_PATH)

# --- 2. Define Features and Target (Predict Weight) ---
target_col = 'Weight'
feature_cols = ['Age', 'Height', 'Gender', 'Smoking_Status']
df_model = df[feature_cols + [target_col]].dropna()

# --- 3. Preprocessing (One-Hot Encoding for Categorical Features) ---
# drop_first=True creates 'Gender_Male', 'Gender_Other', etc.
df_processed = pd.get_dummies(df_model, columns=['Gender', 'Smoking_Status'], drop_first=True)
y = df_processed[target_col]
X = df_processed.drop(columns=[target_col])

# --- 4. Split Data (80% Train, 20% Test) ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# --- 5. Model Training (NO SCALING, to match your text output coefficients) ---
model = LinearRegression().fit(X_train, y_train)

# --- 6. Evaluation ---
# Training Set
y_train_pred = model.predict(X_train)
train_mse = mean_squared_error(y_train, y_train_pred)
train_rmse = np.sqrt(train_mse)
train_mae = mean_absolute_error(y_train, y_train_pred)
train_r2 = r2_score(y_train, y_train_pred)

# Test Set
y_test_pred = model.predict(X_test)
test_mse = mean_squared_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)
test_mae = mean_absolute_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)

# --- 7. Coefficient Formatting ---
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
}).sort_values(by='Coefficient', ascending=False)

# --- 8. Prediction Scenarios (Simulated Patients) ---
required_features = X_train.columns
scenarios_data = pd.DataFrame(0, index=range(4), columns=required_features)
scenarios_data['Age'] = [40, 70, 50, 60]
scenarios_data['Height'] = [170, 185, 160, 175]

# Scenarios set up based on the dummy variables:
# Features: Age, Height, Gender_Male, Gender_Other, Smoking_Status_Non-Smoker, Smoking_Status_Smoker
# Baselines: Gender_Female, Smoking_Status_Former Smoker

# Scenario 1: Non-Smoking Male, Young/Avg Height
scenarios_data.loc[0, 'Gender_Male'] = 1
scenarios_data.loc[0, 'Smoking_Status_Non-Smoker'] = 1

# Scenario 2: Smoking Male, Older/Tall
scenarios_data.loc[1, 'Gender_Male'] = 1
scenarios_data.loc[1, 'Smoking_Status_Smoker'] = 1

# Scenario 3: Smoking Female, Avg Age/Short (Female is baseline)
scenarios_data.loc[2, 'Smoking_Status_Smoker'] = 1

# Scenario 4: Non-Smoking Female, Older/Avg Height (Female is baseline)
scenarios_data.loc[3, 'Smoking_Status_Non-Smoker'] = 1

predicted_weights = model.predict(scenarios_data)

scenario_output = pd.DataFrame({
    'Scenario': ['Non-Smoking Male, Young/Avg Height', 'Smoking Male, Older/Tall', 'Smoking Female, Avg Age/Short', 'Non-Smoking Female, Older/Avg Height'],
    'Age': scenarios_data['Age'],
    'Height': scenarios_data['Height'],
    'Predicted_Weight': predicted_weights
})

# --- 9. Sample Predictions vs Actual (Test Set) ---
comparison = pd.DataFrame({
    'Actual': y_test.head(8),
    'Predicted': y_test_pred[:8]
}).reset_index(drop=True)

comparison['Difference'] = comparison['Actual'] - comparison['Predicted']
comparison['Percentage_Error'] = (comparison['Difference'] / comparison['Actual']) * 100


# --- 10. Print output in the exact requested format ---
print("Cancer Patient Weight Prediction using Multiple Linear Regression")
print("============================================================\n")
print(f"Target variable: {target_col}")
print(f"Features used: {list(X.columns)}")
print(f"Total records (clean): {len(df_processed)}")
print(f"Training set: ({len(X_train)}, {X_train.shape[1]}), Test set: ({len(X_test)}, {X_test.shape[1]})\n")

print("Training Set Evaluation:")
print(f"Mean Squared Error (MSE): {train_mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {train_rmse:.4f}")
print(f"Mean Absolute Error (MAE): {train_mae:.4f}")
print(f"R Score: {train_r2:.4f}\n")

print("Test Set Evaluation:")
print(f"Mean Squared Error (MSE): {test_mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {test_rmse:.4f}")
print(f"Mean Absolute Error (MAE): {test_mae:.4f}")
print(f"R Score: {test_r2:.4f}\n")

print("Model Coefficients:")
print(coefficients.to_string(index=False, float_format='%.4f'))
print(f"\nIntercept: {model.intercept_:.4f}")

print("\n==================================================")
print("PREDICTION SCENARIOS (SIMULATED PATIENTS)")
print("==================================================\n")
print(scenario_output[['Scenario', 'Age', 'Height', 'Predicted_Weight']].to_string(index=False, float_format='%.4f'))

print("\n==================================================")
print("ADDITIONAL ANALYSIS")
print("==================================================\n")

print("Sample Predictions vs Actual (Test Set):")
print(f"{'Actual':>10}{'Predicted':>12}{'Difference':>12}{'Percentage_Error':>20}")
for index, row in comparison.iterrows():
    print(f"{row['Actual']:10.4f}{row['Predicted']:12.4f}{row['Difference']:12.4f}{row['Percentage_Error']:20.4f}")

print("\nModel Interpretation:")
print(f"- R of {test_r2:.4f} indicates the model explains {test_r2*100:.2f}% of variance in Weight.")
print(f"- Average prediction error (MAE): {test_mae:.4f} kg.")
print("- Most important features (by coefficient magnitude):")
top_abs_features = coefficients.reindex(coefficients['Coefficient'].abs().sort_values(ascending=False).index).head(3)
for index, row in top_abs_features.iterrows():
    print(f"  {row['Feature']}: {row['Coefficient']:.4f}")


# --- 11. PLOTTING FUNCTION (The 2x2 Grid) ---
def plot_results(y_test, y_pred_test, test_r2):
    """Generates and displays the 2x2 grid of regression diagnostic plots."""
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    if not isinstance(y_test, pd.Series):
        y_test = pd.Series(y_test)
        
    residuals = y_test - y_pred_test
    
    # 1. Actual vs Predicted Plot (Top Left)
    sns.scatterplot(x=y_test, y=y_pred_test, alpha=0.6, ax=axes[0, 0])
    min_val = min(y_test.min(), y_pred_test.min())
    max_val = max(y_test.max(), y_pred_test.max())
    axes[0, 0].plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Perfect Prediction Line')
    axes[0, 0].set_xlabel('Actual Weight (kg)', fontsize=12)
    axes[0, 0].set_ylabel('Predicted Weight (kg)', fontsize=12)
    axes[0, 0].set_title(f'1. Actual vs Predicted | R={test_r2:.4f}', fontsize=14)
    axes[0, 0].grid(True, linestyle='--')
    
    # 2. Residuals Plot (Top Right)
    sns.scatterplot(x=y_pred_test, y=residuals, alpha=0.6, ax=axes[0, 1])
    axes[0, 1].axhline(y=0, color='r', linestyle='--', lw=2)
    axes[0, 1].set_xlabel('Predicted Weight (kg)', fontsize=12)
    axes[0, 1].set_ylabel('Residuals (Actual - Predicted) (kg)', fontsize=12)
    axes[0, 1].set_title('2. Residuals vs Predicted', fontsize=14)
    axes[0, 1].grid(True, linestyle='--')
    
    # 3. Q-Q Plot of Residuals (Bottom Left)
    stats.probplot(residuals, dist="norm", plot=axes[1, 0])
    axes[1, 0].set_title('3. Normal Q-Q Plot of Residuals', fontsize=14)
    axes[1, 0].set_xlabel('Theoretical Quantiles', fontsize=12)
    axes[1, 0].set_ylabel('Residuals', fontsize=12)
    
    # 4. Scale-Location Plot (Bottom Right)
    std_residuals = residuals / np.std(residuals)
    axes[1, 1].scatter(y_pred_test, np.sqrt(np.abs(std_residuals)), alpha=0.6)
    axes[1, 1].set_xlabel('Predicted Weight (kg)', fontsize=12)
    axes[1, 1].set_ylabel('$\\sqrt{|Standardized\ Residuals|}$', fontsize=12)
    axes[1, 1].set_title('4. Scale-Location Plot (Homoscedasticity Check)', fontsize=14)
    axes[1, 1].grid(True, linestyle='--')
    
    plt.tight_layout()
    
    # *** THIS IS THE KEY CHANGE FOR JUPYTER/ANACONDA ***
    plt.show() # This line will display the plot inline in your notebook
    
    # We remove or comment out savefig and close
    # plt.savefig('mlr_full_analysis_diagnostics_2x2.png')
    # plt.close()

# --- 12. Call the plotting function ---
plot_results(y_test, y_test_pred, test_r2)
print("\n2x2 Regression Diagnostics plots have been displayed inline.")
```

```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Load full dataset
df = pd.read_csv('_cancer_dataset_uae.csv')

# Keep only first 300 rows
df_small = df.head(300)

# Numeric columns only
numeric_df = df_small.select_dtypes(include=['number']).dropna()

# Perform hierarchical clustering
Z = linkage(numeric_df, method='ward')

plt.figure(figsize=(10, 5))

# Truncate dendrogram to avoid overload
dendrogram(
    Z,
    truncate_mode='lastp',  # show only last p merged clusters
    p=12,                   # number of visible clusters
    leaf_rotation=90.,
    leaf_font_size=10.,
    show_contracted=True
)

plt.title("Dendrogram (First 300 Rows of UAE Cancer Dataset)")
plt.xlabel("Cluster Groups")
plt.ylabel("Distance")
plt.tight_layout()
plt.show()
```
